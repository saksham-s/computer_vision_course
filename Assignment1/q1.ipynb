```
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```
image = cv2.imread('image_1.jpg')
```

```
image=image[..., ::-1]/255
plt.imshow(image)
plt.grid(False)
plt.xticks([])
plt.yticks([])
```

```
filter_size = 15
padding = int((filter_size-1)/2)
```

```
padded_image = np.zeros((image.shape[0] + 2*padding, image.shape[1] + 2*padding, image.shape[2]))
padded_image[padding: image.shape[0] + padding, padding: image.shape[1] + padding, :] = image

new_image = np.zeros(image.shape)
```

```
for channel in range(image.shape[2]):
    for i in range(padding,image.shape[0] + padding):
        for j in range(padding,image.shape[1] + padding):
            new_image[i-padding][j-padding][channel] = np.sum(padded_image[i-padding:i+padding + 1, j-padding:j+padding + 1, channel])/(filter_size*filter_size)
```

```
plt.imshow(new_image)
plt.grid(False)
plt.xticks([])
plt.yticks([])
```

```
cv_blurred = cv2.blur(image,(15,15))
```

```
plt.imshow(cv_blurred)
plt.grid(False)
plt.xticks([])
plt.yticks([])
```

```
a = np.abs(new_image - cv_blurred)
print(np.max(a))
plt.imshow(a)
plt.grid(False)
plt.xticks([])
plt.yticks([])
```
